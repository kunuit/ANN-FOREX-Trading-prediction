```python
series = df[["Volume","Open","High","Low","Close"]]
series.shape
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import sklearn.preprocessing as prep
```

```python
def standard_scaler(X_train, X_test):
    train_samples, train_nx, train_ny = X_train.shape
    test_samples, test_nx, test_ny = X_test.shape
    
    X_train = X_train.reshape((train_samples, train_nx * train_ny))
    X_test = X_test.reshape((test_samples, test_nx * test_ny))
    
    preprocessor = prep.StandardScaler().fit(X_train)
    X_test = preprocessor.transform(X_test)
    
    X_test = X_test.reshape((test_samples, test_nx, test_ny))
    
    return  X_test

def preprocess_data(stock, seq_len):
    amount_of_features = len(stock.columns)
    data = stock.values
    
    sequence_length = seq_len + 1
    result = []
    for index in range(len(data) - sequence_length):
        result.append(data[index : index + sequence_length])
        
    result = np.array(result)
    
    X_train = result[:, : -1]
    y_train = result[:, -1][: ,-1].reshape(-1, 1) #y_train s dng 'close' ca ngy d on

    
    #Normalize X
    row = round(0.9 * X_train.shape[0])
    train = X_train[: int(row), :]    
    X_train = standard_scaler(train, X_train)

  
    #Normalize Y
    row = round(0.9 * y_train.shape[0])
    train = y_train[: int(row)]    
    #y_train,preprocessor = standard_scaler(train, y_train)
    preprocessor = prep.StandardScaler()
    preprocessor.fit(train)
    y_train = preprocessor.transform(y_train)

    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], amount_of_features))

    return [X_train, y_train,preprocessor]
```

```python
T = 10
X_train, y_train, preprocessor = preprocess_data(series, T)
print("X_train", X_train.shape)
print("y_train", y_train.shape)
```

```python
plt.plot(y_train)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.20, random_state=2021, shuffle= False)
print("X_train.shape",  X_train.shape, "y_train.shape",y_train.shape,"X_test.shape", X_test.shape,"y_test.shape", y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.recurrent import LSTM
import time, math
```

```python
def build_model(layers):
    model = Sequential()

    # By setting return_sequences to True we are able to stack another LSTM layer
    model.add(LSTM(
        input_dim=layers[0],
        units=layers[1],
        return_sequences=False))

    # model.add(LSTM(
    #     layers[2],
    #     return_sequences=False))

    model.add(Dense(
        units=layers[3]))
    model.add(Activation("linear"))

    start = time.time()
    model.compile(loss="mse", optimizer="rmsprop", metrics=['accuracy'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```python
model = build_model([X_train.shape[2], 9,100, 1])
```

```python
model.fit(
    X_train,
    y_train,
    epochs=50,
    validation_split=0.1,
    verbose=0)
```

```python
trainScore = model.evaluate(X_train, y_train, verbose=0)
print('Train Score: %.2f MSE (%.2f RMSE)' % (trainScore[0], math.sqrt(trainScore[0])))

testScore = model.evaluate(X_test, y_test, verbose=0)
print('Test Score: %.2f MSE (%.2f RMSE)' % (testScore[0], math.sqrt(testScore[0])))
```

```python
pred = model.predict(X_test)
pred.shape
```

```python

predictions = preprocessor.inverse_transform(pred)
# y_test_scaled = preprocessor.inverse_transform(y_test.reshape(-1, 1))
y_test_scaled = preprocessor.inverse_transform(y_test)

fig, ax = plt.subplots(figsize=(16,8))
ax.set_facecolor('#000041')
ax.plot(y_test_scaled[:,0], color='red', label='Original close price')
plt.plot(predictions[:,0], color='cyan', label='Predicted close price')
plt.legend()
```
